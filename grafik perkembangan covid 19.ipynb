```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import requests
import json
```

```python
df = pd.read_json('https://data.covid19.go.id/public/api/update.json')
r = requests.get('https://data.covid19.go.id/public/api/update.json')
r
```

```python
data = r.json()
data
```

```python
daily= data['update']['harian']
daily
```

```python
positive = []
sembuh = []
death = []
date = []
for data in daily:
    tanggal, jam = data['key_as_string'].split("T")
    tahun, bulan, hari = tanggal.split("-")
    if tahun == "2020":
        positive.append(data['jumlah_positif_kum']['value'])
        sembuh.append(data['jumlah_sembuh_kum']['value'])
        death.append(data['jumlah_meninggal_kum']['value'])
        date.append(tanggal)
    
```

```python
positive, sembuh, death
```

```python
semua_data = zip(date, positive, sembuh, death)
semua_data
```

```python
semua_data = list(zip(date, positive, sembuh, death))
semua_data 
```

```python
data_covid = pd.DataFrame(semua_data, columns = ['TANGGAL','Positif','Sembuh','Meninggal'])
data_covid
```

```python
data_covid.plot(x= 'TANGGAL', y=['Positif','Sembuh','Meninggal'], color= ['r','g','black'])
sns.set()
sns.set_style('white')
plt.xticks(rotation=45)
plt.title("Grafik Kasus Meninggal COVID 19 di Indonesia")
plt.show()
```
